```python
import pandas as pd
```

```python
data = pd.read_excel('D:\\Vendor_dataset.xlsx')
```

```python
vendor_data = data.drop(['Question5','Question6','Question7','Question14','Question15','Question16','Question17','Question18','Question19','INSERTEDAT','UPDATEDAT'],axis=1)


```

```python
vendor_data.head()
```

```python
vendor_data['Performance_Score'] = ((vendor_data['Question1']*15/100)+(vendor_data['Question2']*15/100)+(vendor_data['Question3']*9/100)+(vendor_data['Question4']*9/100)+(vendor_data['Question8']*1.5/100)+(vendor_data['Question9']*13/100)+(vendor_data['Question10']*13/100)+(vendor_data['Question12']*11/100)+(vendor_data['Question13']*5.5/100)+(vendor_data['Question11']*8/100))
```

```python
vendor_data.set_index("DATE",inplace = True)
```

```python
Blackwood_Data = vendor_data.loc[vendor_data['VENDORNAME'] == 'BLACKWOODS']
```

```python
Blackwood_Data = Blackwood_Data[['Performance_Score']]
```

```python
Blackwood_Data 
```

```python
Blackwood_Data.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(Blackwood_Data, model ='additive', period = int(len(Blackwood_Data)/2))
seasonal_decompose = result.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
ACF_PLOT = plot_acf(Blackwood_Data)
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
PACF_PLOT = plot_pacf(Blackwood_Data)
```

```python
from statsmodels.tsa.stattools import adfuller
adf, pval, usedlag, nobs, crit_vals, icbest = adfuller(Blackwood_Data.Performance_Score.values)
print('ADF test statistis:', adf)
print('ADF p-values', pval)
print('ADF number of lags used:', usedlag)
print('ADF number of observations:', nobs)
print('ADF critical values:', crit_vals)
print('ADF best information criterion:', icbest)
```

```python
prev_Performance_Score_value = Blackwood_Data.Performance_Score.shift()
differenced_Performance_Score = Blackwood_Data.Performance_Score - prev_Performance_Score_value
differenced_Performance_Score.plot()
```

```python
adf, pval, usedlag, nobs, crit_vals, icbest = adfuller(differenced_Performance_Score.dropna())
print('ADF test statistis:', adf)
print('ADF p-values', pval)
print('ADF number of lags used:', usedlag)
print('ADF number of observations:', nobs)
print('ADF critical values:', crit_vals)
print('ADF best information criterion:', icbest)
```

```python
import numpy as np
```

```python
# extract the seasonality data
months = [x.month for x in Blackwood_Data.index]
years = [x.year for x in Blackwood_Data.index]
day = [x.day for x in Blackwood_Data.index]

#convert into one matrix
X = np.array([day,months,years]).T
```

```python
print(X.shape)
print(Blackwood_Data.Performance_Score.values.shape)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#fit the model
my_lr = LinearRegression()
my_lr.fit(X, Blackwood_Data.Performance_Score.values)

#predict on the same period 
preds = my_lr.predict(X)

#plot what has been learned 
plt.plot(Blackwood_Data.index, Blackwood_Data.Performance_Score.values)
plt.plot(Blackwood_Data.index, preds)
```

```python
from sklearn.ensemble import RandomForestRegressor

#fit the model
my_rf = RandomForestRegressor()
my_rf.fit(X, Blackwood_Data.Performance_Score.values)

#predict on the same period 
preds = my_rf.predict(X)

#plot what has been learned 
plt.plot(Blackwood_Data.index, Blackwood_Data.Performance_Score.values)
plt.plot(Blackwood_Data.index, preds)
```

```python
import xgboost as xgb

#fit the model
my_xgb = xgb.XGBRegressor()
my_xgb.fit(X, Blackwood_Data.Performance_Score.values)

#predict on the same period 
preds = my_xgb.predict(X)

#plot what has been learned 
plt.plot(Blackwood_Data.index, Blackwood_Data.Performance_Score.values)
plt.plot(Blackwood_Data.index, preds)
```
